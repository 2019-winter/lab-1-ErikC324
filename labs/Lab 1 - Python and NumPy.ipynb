## Exercise 6

```python
def random_arrays(low, high):
    for i in range(3):
        rows = np.random.randint(low, high)
        cols = np.random.randint(low, high)
        a = np.array([[np.random.randint(low, high) for i in range(rows)] for j in range(cols)])
        print(a)
        print("Sum: %d"%np.sum(a))
        print("Mean: %f"%np.mean(a))
        print("")

random_arrays(5, 15)
```

## Exercise 7

```python
def random_array(low, high):
    rows = np.random.randint(low, high)
    cols = np.random.randint(low, high)
    a = np.array([[np.random.randint(low, high) for i in range(rows)] for j in range(cols)])
    print(a)
    return a

def one_counter(array):
    count = 0
    for i in range(array.shape[0]):
        for j in range(array.shape[1]):
            if array[i][j] == 1:
                count += 1
    print(count)
    
def one_counter2(array):
    print(np.where(array == 1, 1, 0).sum().sum())
    
arr = random_array(1,6)
one_counter(arr)
one_counter2(arr)
```

## Exercise 11
Repeat exercise A.7 using a dataframe.

```python
def random_df(low, high):
    rows = np.random.randint(low, high)
    cols = np.random.randint(low, high)
    a = pd.DataFrame([[np.random.randint(low, high) for i in range(rows)] for j in range(cols)])
    display(a)
    return a

def one_counter(df):
    count = 0
    for i in range(df.shape[0]):
        for j in range(df.shape[1]):
            if df.at[i, j] == 1:
                count += 1
    print(count)
    
def one_counter2(df):
    print((df == 1).sum().sum())
    
df = random_df(1,6)
one_counter(df)
one_counter2(df)
```
